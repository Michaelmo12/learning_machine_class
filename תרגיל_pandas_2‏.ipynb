```python
import matplotlib.pyplot as plt # Import matplotlib for plotting

# Assuming 'df' is your DataFrame and contains columns 'student_age' and 'hours_of_study'
plt.figure(figsize=(8, 6))  # Adjust figure size if needed
plt.scatter(df['student_age'], df['hours_of_study'])  # Create a scatter plot
plt.xlabel("student_age")  # X-axis label in Hebrew
plt.ylabel("hours_of_study")  # Y-axis label in Hebrew
plt.title("גיל הסטודנט מול סך השעות שלמד")  # Plot title in Hebrew
plt.grid(True)  # Add grid lines
plt.show()  # Display the plot
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Import for 3D plotting

# Filter data for students who studied at least 3 hours
filtered_df = df[df['hours_of_study'] >= 3]

# Create the 3D plot
fig = plt.figure(figsize=(10, 8))  # Adjust figure size if needed
ax = fig.add_subplot(111, projection='3d')  # Create a 3D subplot

# Plot the data
ax.scatter(filtered_df['student_age'], filtered_df['hours_of_study'], filtered_df['grade'])

# Set labels (Hebrew)
ax.set_xlabel("student_age", fontsize=12)
ax.set_ylabel("hours_of_study", fontsize=12)
ax.set_zlabel("grade", fontsize=12)
ax.set_title("גיל, שעות לימוד וציון (לסטודנטים שלמדו לפחות 3 שעות)", fontsize=14)

plt.show()
```

```python
avg_grade = df['grade'].mean()

# Create the 'grade_improve' column
df['grade_improve'] = df['grade'].apply(lambda grade: avg_grade if grade < avg_grade else grade)

# Display the updated DataFrame
df
```
